```python
from collections import namedtuple
import numpy as np
import matplotlib.pyplot as plt

import pyga

from star import Star
```

```python
Point = namedtuple("Point", ["x", "y"])

def get_star_flat():
    outer_radius = 1.0
    inner_radius = 0.5
    num_points = 10
    star_points = []

    for i in range(num_points):
        angle = i * (2 * np.pi / num_points)
        radius = outer_radius if i % 2 == 0 else inner_radius
        x = radius * np.cos(angle)
        y = radius * np.sin(angle)
        star_points.extend([x, y])

    return star_points

# Set target shape once
Star.set_target_shape(get_star_flat())
```

```python
population_size = 25
individuals = [Star(np.random.uniform(-1, 1, size=20)) for _ in range(population_size)]
population = pyga.Population(individuals)
population.plot()
```

```python
ga = pyga.GeneticAlgorithm(population, 
                           num_generations=100, 
                           num_parents=4, 
                           mutation_probability=0.05, 
                           animate=True)
```

```python
ga.evolve()
```

```python
ga.plot_fitness()
```

```python
ga.population.plot()
```
